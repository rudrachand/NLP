```python
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

df1['avg_word'] = df1['Description'].apply(lambda x: avg_word(x))
#df1[['Description','avg_word']].head()
```

# Basic Pre-processing

1)Lower Case

2)Removing Punctuation

3)Removal of Stop Words

4)Common word removal

5)Rare words removal

6)Spelling correction

7)Tokenization

8)Stemming

9)Lemmatization


```python
#Lower Case
df1['Description'] = df1['Description'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df1['Description'].head()
```

```python
#Removing Punctuation
df1['Description'] = df1['Description'].str.replace('[^\w\s]','')
df1['Description'].head()
```

```python
#Removal of Stop Words
from nltk.corpus import stopwords
stop = stopwords.words('english')
df1['Description'] = df1['Description'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df1['Description'].head()
```

```python
#Common word removal
#let’s check the 10 most frequently occurring words in our text data 
freq = pd.Series(' '.join(df1['Description']).split()).value_counts()[:10]
```

```python
freq
```

```python
#Now, let’s remove these words as their presence will not of any use in classification of our text data.
freq = list(freq.index)
df1['Description'] = df1['Description'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
df1['Description'].head()
```

```python
#Rare words removal
#remove rarely occurring words from the text.
freq = pd.Series(' '.join(df1['Description']).split()).value_counts()[-10:]
```

```python
freq
```

```python
freq = list(freq.index)
df1['Description'] = df1['Description'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
df1['Description'].head()
```

```python
#Spelling correction
from textblob import TextBlob
df1['Description'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
#Tokenization
#Tokenization refers to dividing the text into a sequence of words or sentences
TextBlob(df1['Description'][1]).words
```

```python
#Stemming
#Stemming refers to the removal of suffices, like “ing”, “ly”, “s”, etc. 
from nltk.stem import PorterStemmer
st = PorterStemmer()
df1['Description'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
#Lemmatization
#Lemmatization is a more effective option than stemming because it converts the word into its root word
from textblob import Word
df1['Description'] = df1['Description'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df1['Description'].head()
```

# Advance Text Processing
1) N-grams

2) Term frequency

3) Inverse Document Frequency

4) Term Frequency – Inverse Document Frequency (TF-IDF)

5) Bag of Words

6) Sentiment Analysis

7) Word Embeddings

```python
#N-grams
# N-grams are the combination of multiple words used together. 
# Ngrams with N=1 are called unigrams. Similarly, bigrams (N=2), trigrams (N=3) and so on can also be used.
TextBlob(df1['Description'][0]).ngrams(2)
```

```python
# Term Frequency
# TF = (Number of times term T appears in the particular row) / (number of terms in that row)
tf1 = (df1['Description'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

```python
# Inverse Document Frequency
# The intuition behind inverse document frequency (IDF) is that a word is not of much use to us if it’s appearing in all the documents.

# Therefore, the IDF of each word is the log of the ratio of the total number of rows to the number of rows in which that word is present.
for i,word in enumerate(tf1['words']):
    tf1.loc[i, 'idf'] = np.log(df1.shape[0]/(len(df1[df1['Description'].str.contains(word)])))
tf1    
```

```python
#TF-IDF
# TF-IDF is the multiplication of the TF and IDF which we calculated above.
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```

```python
#We don’t have to calculate TF and IDF every time beforehand and then multiply it to obtain TF-IDF.
# Instead, sklearn has a separate function to directly obtain it
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1000, lowercase=True, analyzer='word',
 stop_words= 'english',ngram_range=(1,1))
DESCRIPTION_vect = tfidf.fit_transform(df1['Description'])
DESCRIPTION_vect
```

```python
#Bag of Words
# Bag of Words (BoW) refers to the representation of text which describes the presence of words within the text data
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer(max_features=1000, lowercase=True, ngram_range=(1,1),analyzer = "word")
DESCRIPTION_bow = bow.fit_transform(df1['Description'])
DESCRIPTION_bow
```

```python
#Sentiment Analysis
df1['Description'][:5].apply(lambda x: TextBlob(x).sentiment)
```

```python
#Above, you can see that it returns a tuple representing polarity and subjectivity of each Description. 
df1['sentiment'] = df1['Description'].apply(lambda x: TextBlob(x).sentiment[0] )
df1[['Description','sentiment']].head()
```
